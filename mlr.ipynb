```
import numpy as np
x = np.array([[1,15,0.75,55],
              [1,26,0.8,45],
              [1,16,0.85,75],
               [1,18,0.88,50],
               [1,30,0.94,65]])
y= [[55],
    [68],
    [62],
    [65],
    [72]]


print(x)

xt = x.transpose()
a= xt.dot(x)
ainv = np.linalg.inv(a)
b= ainv.dot(xt)
beta= b.dot(y)

print(beta)
```
