```
tc = int(input())

for t in range(1, tc+1):
  b1 =[]
  b2= []
  line =input().split()

  n= int(line[0])
  k= int(line[1])


  for i in range(0,k):
    b1.insert(i,0)
    b2.insert(i,0)

 

  while(sum(b1)!= n):
    for i in range(0,k):
      b1[i] = b1[i]+1


  while(sum(b2)!= n):
    for i in range(0,k):
      if sum(b2)!= n:
        b2[i]= b2[i] + k
      elif sum(b2) == n:
        break

  if b1== b2 :
    print("YES")
  elif b1 != b2:
    print("NO")






  

```
